# Correlation matrix

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load from the file
processed_data = pd.read_csv('Data/Processed_Data.csv')
processed_data.columns = ['Возраст',

                          'Женский', 'Мужской',

                          'Бакалавриат 1 курс', 'Бакалавриат 2 курс', 'Бакалавриат 3 курс', 'Бакалавриат 4 курс', 'Магистратура',

                           'Архитектура', 'Бизнес', 'Другое', 'Инженерия', 'Програмирование', 'Психология', 'Юриспруденция',

                           'Экстраверсия – интроверсия', 'Привязанность - обособленность', 'Самоконтроль - импульсивность', 'Эмоциональная устойчивость -\n эмоциональная неустойчивость', 'Экспрессивность - практичность', 'Общий балл по факторам',

                           'Проблемно-ориентированная', 'Эмоционально-ориентированная', 'Ориентированная на избегание', 'Отвлечение', 'Поиск социальной поддержки', 'Лучшая стратегия', 'Общий балл по стратегиям',

                           'Общий результат']
processed_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the correlation matrix
corr = processed_data.corr(method='pearson')
plt.figure(figsize=(17, 12))

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='gray', cbar=True, square=True, fmt=".2f", xticklabels=corr.columns, yticklabels=corr.columns, center=0 , vmin=-1, vmax=1, mask=np.triu(corr))

plt.axhline(y=15, color='black', linewidth=2)
plt.axhline(y=21, color='black', linewidth=2)
plt.axhline(y=28, color='black', linewidth=2)

plt.axvline(x=15, color='black', linewidth=2)
plt.axvline(x=21, color='black', linewidth=2)
plt.axvline(x=28, color='black', linewidth=2)

# add bracket with labels for groups of columns (factors, strategies, etc)
plt.text(-8.5, 8, 'Персональные\n характеристики', fontsize=13, color='black', ha='center', va='center')
plt.text(-9.5, 18, '5PFQ', fontsize=13, color='black', ha='center', va='center')
plt.text(-10.5, 24.5, 'Копинг-\nСтратегии', fontsize=13, color='black', ha='center', va='center')


# draw line over the plot
plt.plot([-10, 0], [0, 0], color='black', linewidth=2)



plt.title('Корреляционная матрица переменных (коэффициент Пирсона)', fontsize=15, color='black')

plt.savefig("Illustrations/Corr_heat_map.png", dpi=200, bbox_inches = 'tight')
plt.show()
```
